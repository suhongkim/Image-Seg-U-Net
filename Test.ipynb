```python
%matplotlib inline
import sys
import os
from os.path import join
from optparse import OptionParser
import numpy as np

import torch
import torch.backends.cudnn as cudnn
import torch.nn as nn
from torch import optim

from torchvision import transforms

import matplotlib.pyplot as plt

from model import UNet
from dataloader import DataLoader
from train import train_net, get_args
```

```python
# Show the best model
loadModel = True
model_path = './data/cells/checkpoints/CP_best.pth'

gpu = torch.cuda.is_available()
epochs = 0
n_classes = 2
data_dir = 'data/cells/'
lr = 0.001
net = UNet(n_classes=n_classes)

if loadModel:
    net.load_state_dict(torch.load(model_path))
    print('Model loaded from %s' % (model_path))

if gpu:
    net.cuda()
    cudnn.benchmark = True

train_net(net=net, epochs=epochs, lr=lr, n_classes=n_classes, gpu=gpu, data_dir=data_dir)
```

```python
# For Training 

loadModel = False

gpu = torch.cuda.is_available()
epochs = 5
n_classes = 2
data_dir = 'data/cells/'
lr = 0.001

test_data = epochs

net = UNet(n_classes=n_classes)

if loadModel:
    net.load_state_dict(torch.load(model_path))
    print('Model loaded from %s' % (model_path))

if gpu:
    net.cuda()
    cudnn.benchmark = True

train_net(net=net, epochs=epochs, lr=lr, n_classes=n_classes, gpu=gpu, data_dir=data_dir)
```
